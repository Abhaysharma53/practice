```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

## making a simple plot

```python
# list of height
height = [150,160,165,185]
# list of weight
weight = [70, 80, 90, 100]

plt.plot(height, weight)
#though x is representing height , but from here we cant explicitly tell this
```

```python
plt.xlabel('height')
plt.ylabel('weight')
plt.title('relation between height and weight')
plt.plot(height, weight)
```

```python
# list of calories_burnt
calories_burnt = [65, 75, 95, 99]
plt.plot(calories_burnt)
plt.plot(weight)
#but from here we wont be able to tell which curve is representing what 
```

```python
plt.plot(calories_burnt)
plt.plot(weight)
plt.legend(labels=['calories burnt', 'weight'], loc = 'lower right');
```

```python
plt.plot(calories_burnt)
plt.plot(weight)
plt.legend(labels=['calories burnt', 'weight'], loc = 'lower right')
plt.xticks(ticks = [0,1,2,3], labels = ['p1', 'p2', 'p3', 'p4']);
```

```python
plt.figure(figsize = (15,5))
plt.plot(calories_burnt, marker = 'o')
plt.plot(weight, marker = '*')
plt.legend(labels=['calories burnt', 'weight'], loc = 'lower right')
plt.xticks(ticks = [0,1,2,3], labels = ['p1', 'p2', 'p3', 'p4']);
```

```python
plt.figure(figsize = (15,5))
plt.plot(calories_burnt,'b--', marker = 'o')
plt.plot(weight, 'r--', marker = '*')
plt.legend(labels=['calories burnt', 'weight'], loc = 'lower right')
plt.xticks(ticks = [0,1,2,3], labels = ['p1', 'p2', 'p3', 'p4']);
#plt.figure(figsize) uses inches as measurement
```

# making subplots

```python
fig, ax = plt.subplots(nrows =2, ncols = 2, figsize = (6,6))
ax[0,0].plot(height)
#ax[0,0].plot(calories_burnt)
ax[0,1].plot(weight, marker = '*')
ax[1,0].plot(calories_burnt)
ax[1,1].plot(calories_burnt, 'go')
ax[0,0].set_title('height curve')
ax[0,1].set_title('weight curve')
ax[1,0].set_title('burnt curve')
ax[0,0].set_xticks(ticks = [0,1,2,3])
ax[0,1].set_xticks(ticks = [0,1,2,3])
ax[0,0].set_xticklabels(labels = ['p1','p2','p3','p4'])
ax[0,0].set_xticklabels(labels = ['p1','p2','p3','p4'])
```

```python
#creating 2 plots with shared axis
fig, ax = plt.subplots(nrows =1, ncols = 2, figsize = (12,6), sharex = True, sharey = True)
ax[0].plot(weight, marker = '*' )
ax[1].plot(calories_burnt, 'go')
ax[0].set_title('weight curve')
ax[1].set_title('calories curve')
ax[0].set_xticks(ticks = [0,1,2,3]);
ax[1].set_xticks(ticks = [0,1,2,3]);
ax[0].set_xticklabels(labels = ['p1', 'p2', 'p3', 'p4']);
ax[0].set_xticklabels(labels = ['p1', 'p2', 'p3', 'p4']);
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(12,6), sharex=True, sharey=True)


# plot on 0 row and 0 column
ax[0].plot(calories_burnt,'go')

# plot on 0 row and 1 column
ax[1].plot(weight)

# set titles for subplots
ax[0].set_title("Calories Burnt")
ax[1].set_title("Weight")

# set ticks for each of these persons
ax[0].set_xticks(ticks=[0,1,2,3]);
ax[1].set_xticks(ticks=[0,1,2,3]);

# set labels for each of these persons
ax[0].set_xticklabels(labels=['p1', 'p2', 'p3', 'p4']);
ax[1].set_xticklabels(labels=['p1', 'p2', 'p3', 'p4']);
```
