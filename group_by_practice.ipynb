```python
ipl = pd.read_csv('/content/deliveries.csv')
ipl.head()
```

```python
# find the top 10 batsman in terms of runs
ipl.groupby('batsman')['batsman_runs'].sum().sort_values(ascending = False).head(10)
```

```python
ipl.shape
```

```python
# find the top 3 batsman with max no of sixes
df = ipl[ipl['batsman_runs'] == 6]
df.groupby('batsman')['batsman_runs'].count().sort_values(ascending= False).head(3)
```

```python
# find batsman with most number of 4's and 6's in last 5 overs
mask1 = ipl['over'] > 15
mask2 = (ipl['batsman_runs'] == 4) | (ipl['batsman_runs'] == 6)
ipl[mask1 & mask2].groupby('batsman')['batsman_runs'].count().sort_values(ascending = False).head()
```

```python
# Create a function that can return the highest score of any batsman
def high_score(batsman):
  temp_df = ipl[ipl['batsman'] == batsman]
  return temp_df.groupby('match_id')['batsman_runs'].sum().sort_values(ascending = False).head(1).values[0].item()
```

```python
high_score('KA Pollard')
```
