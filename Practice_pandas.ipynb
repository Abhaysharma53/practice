filter rows of a pandas dataframe by a column value

```python
#imdb rating
df.shape
```

```python
#lets say we want all the movies with duration more than equal to 200 minutes
booleans = list()
for movies in df.duration:
    if movies >= 200:
        booleans.append(True)
    else:
        booleans.append(False)
```

```python
booleans[0:5]
```

```python
len(booleans)
```

```python
#to convert this to a pandas series
is_long = pd.Series(booleans)
type(is_long)
```

```python
is_long.head()
```

```python
#now result
df[is_long]
#its basicallydisplays the dataframe for all those values for which is_long have True
```

```python
#alternative way without using for loop
is_long = df.duration >= 200
is_long
```

```python
#direct code which let us do the same
df[df.duration >= 200]
#inner part is creating a series of booleans which is being used by outer part to refer to the exact data 
```

```python
#lets say out of this i want to select some specifc rows
#1stway
df[df.duration >= 200].genre
#or
df[df.duration >= 200]['genre']
```

```python
#2nd way using loc
df.loc[df.duration >= 200, 'genre']    #df.loc(rows to select, columns)
```

### applying multiple filter criteria to dataframe

```python
#imdb
df.head(12)
```

```python
#using logical operatior
#in pandas = and => &
# or => |
# lets say we want all the movies with duration >= 200 and genre is crime
df[(df.duration >= 200) & (df.genre == 'Crime')]
```

```python
#lets say we have multiple such condition likke crime, action, drama instead of combining then with multiple pipe operator we can use isin()
df[df.genre.isin(["Crime", "Drama", "Adventure"])]
#similar to in operator of sql
```

### how can i use string methods in pandas

```python
#chip orders dataset
orders = pd.read_csv("http://bit.ly/chiporders", sep = '\t')
orders.head()
```

```python
#to change to upper case alll the item names
orders.item_name.str.upper()
```

```python
orders.head()
```

```python
#lets say i want to filter out all  those item name which contains chicken
orders[orders.item_name.str.contains('Chicken')]
#if we want to use any string method on dataframes, we have to use .str along with it
```

```python
orders.shape
```

```python
#we can chain togethor string methods
#lets say i want to remove these brackets from choice description []
orders['choice_description'] = orders.choice_description.str.replace("[","").str.replace(']','')
orders.head()
```

```python
#can perform this task using regular expersioon even more effectively
orders.choice_description.str.replace('[/[/]]','')   #use / for escape sequence
```

# when should i use group by in pandas?

```python
#drinks by country
drinks.head()
```

```python
drinks.groupby('continent').beer_servings.mean()
```

```python
#lets say our question is what is the avg alcohol consumption by country
drinks.groupby('continent').beer_servings.mean().plot(kind='bar')
```

```python
#if i want to analyse a pandas series by some category i should be using groupby
drinks.groupby('continent').beer_servings.max()
```

```python
#.agg() function let us define multiple aggregation conditions togethor
drinks.groupby('continent').beer_servings.agg(['count', 'min','max', 'mean'])
```

```python
#if we dont specify any column it lets do for all the avaliable columns
drinks.groupby('continent').mean()
```

```python
#can further visualize it
drinks.groupby('continent').mean().plot(kind = 'bar')
```
